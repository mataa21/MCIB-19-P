```python
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Simulando medición de 6 datos
x  = np.arange(6)[:,np.newaxis]
y = np.random.randint(-5,5,(x.shape[0],1))
plt.figure( figsize =(12,4))
plt.plot(x,y,'go', markersize=12)
plt.grid(True)
```

```python
def ones_ext(X):
    X_ext = np.concatenate(  (X,np.ones(  (X.shape[0],1) ) ) , axis=1  )
    return X_ext.copy()

def pinv(X,y):
    w = np.linalg.inv( np.dot( X.T, X ) )
    w = np.dot(w,X.T)
    w = np.dot(w,y)
    return w
```

```python
#Aproximar mediciones con una línea recta
m = pinv (ones_ext(x),y)
ys = np.dot( ones_ext(x),m)
plt.figure( figsize =(12,4))
plt.plot(x,y,'go', markersize=12)
plt.plot(x,ys,'r-', linewidth=3)
plt.grid(True)
plt.show()
```

```python
G = 5

y = np.array([1,1,1,1,0,0])
x_tst = np.linspace([1E-6],[5],100)

x_ext = np.concatenate ( tuple (x   **i for i in range(1,G+1) ) [::-1], axis = 1)
x_tst = np.concatenate (tuple(x_tst**i for i in range(1,G+1) )[::-1], axis=1)

w = pinv    (ones_ext(x_ext),y )
ys = np.dot( ones_ext(x_tst),w )
plt.figure( figsize =(12,8) )
plt.plot(x         , y            , 'go', markersize=12)
plt.plot(x_tst[:,-1],ys.flatten(),'b.', linewidth=3)
plt.grid(True)
plt.show()
```

```python
import wfdb
#Registro de EMG de Physionet
path = '/Users/josec/OneDrive/Documentos/MCIB-19-P/data'
fname = '/e0103'
data = wfdb.rdsamp(path+fname)

samp = 20000
sr = data[1]['fs']
data = data[0][:samp:,0]
shape = data.shape[0]
t= np.linspace(0,shape/sr,shape)

plt.figure( figsize =(12,4))
plt.plot (t,data, color ='green')
plt.grid(True)
plt.figure( figsize=(12,4))
plt.psd(data,Fs=sr, NFFT = 2**10)
plt.show()
print(sr)
```

```python
sub   = 4
data_sub = data[::sub]
t_sub =    t[::sub]
sr_sub   = sr/sub


plt.figure( figsize=(12,4))
plt.plot( t,  data          , color='green')
plt.plot(t_sub, data_sub, '-', color='red')
plt.grid(True)
plt.figure( figsize=(12,4))
plt.psd( data, Fs = sr, NFFT = 2**10, color='green')
plt.psd(data_sub, Fs = sr_sub, NFFT= 2**10, color='red')
```
