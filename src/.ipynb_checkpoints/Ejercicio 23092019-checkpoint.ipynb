```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Declaraci√≥n de constantes
sr = 320
L = 5

F= []
F.append(5)
F.append(30)

```

```python
t = np.linspace(0,L,L*sr)[:,np.newaxis]
F = np.array(F)[np.newaxis] 
```

```python
p = (2*np.pi*t*F)
p = np.sin(p)
p = p.sum(axis=1)+2
p.shape
```

```python
plt.figure(figsize=(10,5))
plt.plot(p[0:300], color = 'green', linewidth = 4)
plt.grid(True)
plt.show()
```

```python
#Filtrado
```

```python
from scipy import signal 
b,a= signal.butter(4,2*15/sr,btype ='high')   
y_f = signal.lfilter(b,a,p)
y_ff = signal.lfilter(b,a,y_f[::-1])[::-1] #Es lo mismo que hace filtfilt
```

```python
plt.figure(figsize=(10,5))
plt.plot(p[0:300], color = 'green', linewidth = 4)
plt.plot(y_f[0:300], color = 'red', linewidth = 4)
plt.plot(y_ff[0:300], color = 'orange', linewidth = 4)
plt.grid(True)
plt.show()
```

```python
N = 40
b = signal.firwin(N, 2/sr*15)
y_f = signal.lfilter(b,[1],p)
plt.plot(p[:300], color = 'green', linewidth = 4)
plt.plot(y_f[N//2:300+N//2], color = 'blue', linewidth = 4)
plt.grid(True)
plt.show()
```
