```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
covar =[]
covar.append(np.array([[1,0],[0,1]]))
covar.append(np.array([[1,1],[1,1]]))
covar.append(np.array([[1,-1],[-1,1]]))
covar
```

```python
prom = np.array([0,0])
```

```python
def cova(num):
    if num==0:
        s = np.random.multivariate_normal(prom,covar[num],1000) 
        x = ('Sin correlación')
    elif num==1:
        s = np.random.multivariate_normal(prom,covar[num],1000)
        x =('Correlación directa')
    elif num==2:
        s = np.random.multivariate_normal(prom,covar[num],1000)
        x =('Correlación inversa')
    else:
        print('Número incorrecto')
            
    return s,x
```

```python
print('Presione: 0 sin correlación, 1 correlación directa, 2 correlación inversa')
num = input('Introduce el número \n')
r,x= cova(int(num))
plt.figure(figsize =(12,8))
plt.plot(r[:,0],r[:,1],'*')
plt.title('%s'%x)
plt.show()
```
