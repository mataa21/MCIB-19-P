```python
data.shape
lim = [30,45]
time = (data[:,0]>=lim[0])*(data[:,0]<=lim[1])
time
```

```python
plt.figure(1)
plt.plot(data[time,1])
plt.title("Oxímetro")  
plt.xlabel("s")   
plt.ylabel("magnitud")   
plt.figure(2)
plt.plot(data[time,2])
plt.title("Respirograma")  
plt.xlabel("s")   
plt.ylabel("magnitud")
plt.figure(3)
plt.plot(data[time,3])
plt.title("ECG")  
plt.xlabel("s")   
plt.ylabel("magnitud")
plt.figure(4)
plt.plot(data[time,4])
plt.title("Oxímetro")  
plt.xlabel("s")   
plt.ylabel("magnitud")
plt.figure(5)
plt.plot(data[time,5])
plt.title("Onda de pulso")  
plt.xlabel("s")   
plt.ylabel("magnitud")
```

```python
fs= 1000
```

```python
Tt = len(time)/fs
fs2 = 300
muestras = int(Tt*fs2)
ix = np.arange(1,muestras+1)
ix = np.floor(ix/30)
ix
```

```python
square = np.zeros(muestras)
square[ix%2==0]= 1
square[ix%2!=0]= -1
plt.figure()
plt.plot(square[0:200])
trian=np.cumsum(square)
plt.figure()
plt.plot(trian[0:200])
```
